## Задание 1 
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
five_star_movies = ratings[ratings.rating == 5.0 ]
five_star_movies.head()
```

```python
stats = movies.merge(five_star_movies, on='movieId', how='left')
stats.head()
```

```python
stats['title'].value_counts().index[0]
```

## Задание 2 

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
```

```python
filtered_countries = data[data['country'].isin(['Latvia', 'Estonia', 'Lithuania'])]
filtered_category = filtered_countries[ ((filtered_countries['category']==4) | (filtered_countries['category']==12)) | (filtered_countries['category']==21) ]
```

```python
filtered_year = filtered_category[ (filtered_category['year'] > 2004) & (filtered_category['year'] < 2011) ]
filtered_quantity = filtered_year[ (filtered_year['quantity'] >= 0) ]
filtered_quantity
```

```python
country_sum = filtered_quantity.groupby(['country']).sum()
country_sum = country_sum.filter( items = ['country', 'quantity'])
country_sum
```
