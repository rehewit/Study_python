**Задание 1**
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений


```python
import pandas as pd
import datetime
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log.loc[log.traffic_source.isin(['yandex', 'google']), 'traffic_type'] = 'organic'
log.loc[(log.traffic_source.isin(['email', 'paid'])) & (log.region == 'Russia') , 'traffic_type'] = 'ad'
log.loc[(log.traffic_source.isin(['email', 'paid'])) & (log.region != 'Russia') , 'traffic_type'] = 'other'
log.loc[pd.isnull(log.traffic_type), 'traffic_type'] = log.traffic_source 
log.head()
```

**Задание 3**

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python

>>> str(datetime.timedelta(seconds=666))
```

```python
ratings = pd.read_csv('ratings.csv', sep=',')
users_grouped = ratings.groupby('userId').count().filter(items = ['userId', 'movieId'])
users_grouped.rename(columns={'movieId': 'votes'}, inplace = True)
users_grouped = users_grouped[(users_grouped['votes'] > 100)]
users_grouped.head()
```

```python
agg = ratings[['userId', 'timestamp']].groupby('userId').agg([min, max]).reset_index()
agg['spent_time'] = agg['timestamp', 'max'] - agg['timestamp', 'min']
agg.columns = agg.columns.droplevel(1)
agg.drop(['timestamp'], axis='columns', inplace=True)
user_spent_time = users_grouped.merge(agg, how = 'left', on =['userId'])
str(datetime.timedelta(seconds=(user_spent_time['spent_time'].mean())))
```


В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
```

```python
from pymystem3 import Mystem
m = Mystem()
```

```python
def lemmas(search):
    return ' '.join(m.lemmatize(search)).replace('\n', '') 
```

```python
data['search_result'] = data.keyword.apply(lemmas)
data
```
